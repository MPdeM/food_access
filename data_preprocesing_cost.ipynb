### Cleaning data

Rename columns. Replace symbols. Remove NaN data from FIPS because corresponds to rows with data at the State level.
Remove % symbols and transform to proper numerical formats. 

```python
# rename 
df2 = df2.rename(columns ={'State_and_County_FIPS_Code':'FIPS'})
# replace *
df1 = df1.replace("*","")
```

```python
# remove data NaN at the National and State level
df2 = df2.dropna(subset=['FIPS'])
df2.head()
```

```python
for (columnName, columnData) in df2.iteritems():
    if columnName != 'State' and columnName != 'County':
        splitName = columnName.split("_")
  
        if 'Percent' in splitName:
            df2[columnName] = df2[columnName].str.split(" ", n = 1, expand = True)[0]
        elif 'Rate' in splitName:
            df2[columnName] = df2[columnName].str.split(" ", n = 1, expand = True)[0]
        elif '%' in splitName:
            df2[columnName] = df2[columnName].str.split(" ", n = 1, expand = True)[0]

        df2[columnName] = pd.to_numeric(df2[columnName], errors='coerce')


df2.head(20)
```
